```python
# Load the transaction data
df1 = pd.read_csv('/content/drive/MyDrive/Fraud Dataset/transaction_dataset_corrected (1) (2).csv')
df1.head()
```

```python
#load customer data
df2 = pd.read_csv('/content/drive/MyDrive/Fraud Dataset/customer_dataset 1.csv')
df2.head()
```

```python
# Merge the datasets
merged_data = pd.merge(df1, df2, left_on='Sender Account', right_on='Customer ID')
```

```python
merged_data.head(10)
```

```python
df = merged_data.drop(['Transaction ID','Name','Timestamp'], axis=1)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split


# Split the dataset into training and testing sets
df_train, df_test = train_test_split(df, test_size=0.2)

# Save the training set as a CSV file
df_train.to_csv("training_set.csv", index=False)

# Save the testing set as a CSV file
df_test.to_csv("testing_set.csv", index=False)
```

```python
import h2o
from h2o.automl import H2OAutoML
h2o.init()
```

```python
# Convert pandas DataFrame to H2OFrame
h2o_frame = h2o.H2OFrame(df_train)

# Specify the target column and ensure it's categorical
y = 'Fraud'
h2o_frame[y] = h2o_frame[y].asfactor()

# Define feature columns by removing the target column
x = h2o_frame.columns
x.remove(y)
```

```python
h2o_automl = H2OAutoML(sort_metric='mse', max_runtime_secs=5*60, seed=666)
h2o_automl.train(x=x, y=y, training_frame=h2o_frame)
```

```python
h2o_models = h2o.automl.get_leaderboard(h2o_automl, extra_columns = "ALL")
h2o_models
```

```python
from h2o.model.regression import h2o_mean_squared_error

h2o_frame_test = h2o.H2OFrame(df_test)

y_pred = h2o_automl.predict(h2o_frame_test)
# Extract the prediction column corresponding to the '1' class (assuming 'Fraud' is 1/0)
# or the relevant class label if it's a multi-class problem
y_pred = y_pred['p1']  # Replace 'p1' with the appropriate column name if needed

y_actual = h2o.H2OFrame(df_test[['Fraud']])

h2o_mean_squared_error(y_actual, y_pred)
```

```python
y_pred = y_pred.as_data_frame()
y_actual = y_actual.as_data_frame()
result = pd.concat([y_pred, y_actual], axis=1)
result
```
